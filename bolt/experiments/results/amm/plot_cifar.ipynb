```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cifar10 = pd.read_csv("cifar10.csv")
cifar100 = pd.read_csv("cifar100.csv")
cifar10.columns
```

```python
cifar100aug = cifar10.copy()
cifar100aug = cifar100aug[(cifar100aug["objective"] != "mse-sklearn")]
cifar100aug = cifar100aug[(cifar100aug["objective"] != "mse-sklearn") & (cifar100aug["objective"] != "mse")]

cifar100aug["objective"][cifar10["objective"].isna()] = ''
cifar100aug["nonzeros_heuristic"][cifar10["nonzeros_heuristic"].isna()] = ''
cifar100aug["method-objective-heuristic"] = cifar100aug['method'].astype(str) +" "+ cifar100aug["objective"] +" "+ cifar100aug["nonzeros_heuristic"]
cifar100aug["objective-heuristic"] = cifar100aug["objective"] +" "+ cifar100aug["nonzeros_heuristic"]

cifar100_exact2 = cifar100aug[cifar100aug.method == "Exact"]
cifar100_exact2.at[0, "ncodebooks"] = 2
cifar100_exact64 = cifar100aug[cifar100aug.method == "Exact"]
cifar100_exact64.at[0, "ncodebooks"] = 64
cifar100aug = cifar100aug.append(cifar100_exact2)
cifar100aug = cifar100aug.append(cifar100_exact64)
cifar100aug = cifar100aug.reset_index()

fig = plt.figure();
g = sns.lineplot(data=cifar100aug, x='ncodebooks', y='acc_amm', hue='method-objective-heuristic')
g.set(xscale='log')
g.set(xticks=cifar100['ncodebooks'].dropna().unique(), xticklabels=cifar100['ncodebooks'].dropna().unique());

fig = plt.figure(figsize=(5, 5),dpi=150)
markers = {
    'Mithral  ': '*', 'Exact  ': '*',
    'Pluto kld pq': '*', 'Pluto kld r2': 'v', 'Pluto kld opq': 'P',
    'Pluto mse pq': '*', 'Pluto mse r2': 'v', 'Pluto mse opq': 'P',
    #'Pluto mse-sklearn pq': '*', 'Pluto mse-sklearn r2': 'v', 'Pluto mse-sklearn opq': 'P',
}
dashes = {
    'Mithral  ': (1, 0), 'Exact  ': (1, 0),
    'Pluto kld pq': (1, 0), 'Pluto kld r2': (1, 0), 'Pluto kld opq': (1, 0),
    'Pluto mse pq': (1, 0), 'Pluto mse r2': (1, 0), 'Pluto mse opq': (1, 0),
    #'Pluto mse-sklearn pq': (2, 2, 10, 2), 'Pluto mse-sklearn r2': (2, 2, 10, 2), 'Pluto mse-sklearn opq': (2, 2, 10, 2),
}
g = sns.lineplot(
    data=cifar100aug, x='ncodebooks', y='acc_amm',
    hue='method', hue_order=['Pluto', 'Mithral', 'Exact'],
    style='method-objective-heuristic', markers=markers, markeredgecolor=None,
    dashes=dashes)
g.set(xscale='log')
g.set(
    xticks=cifar100['ncodebooks'].dropna().unique().astype(int),
    xticklabels=cifar100['ncodebooks'].dropna().unique().astype(int));
g.set(title='CIFAR10');
```
